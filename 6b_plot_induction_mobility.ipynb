```python
from os.path import join
from pathlib import Path

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
## 
sheets = ('Flask1', 'Flask2', 'Flask3',) 
folder = r"T:\MIGRATED\Lab_Members\Daniel_Foust\Data\Basal Bodies\fla-che induction\Inducible Water"
fname = r"mobility results summary.xlsx"

data_excel = pd.read_excel(join(folder, fname), sheet_name=None)
data = (data_excel[sheet] for sheet in sheets) # small batch

data = pd.concat(data, ignore_index=True)
data['t_inc'] = data['movie'].str.extract('(\d+)inc').astype('float')
data['t_total'] = data['movie'].str.extract('(\d+)total').astype('float')
data['flask'] = data['movie'].str.extract('Flask(\d+)')

data['% mobile_particles'] = data['f_mobile_particles'] * 100
data['% mobile_particles_weighted'] = data['f_mobile_particles_weighted'] * 100
```

```python
fig_folder = r'T:\MIGRATED\Lab_Members\Daniel_Foust\Data\Basal Bodies\fla-che induction\figures'
fig_savename = r'% bb mobile vs time since induction in water.svg'

pal = sns.color_palette('rocket', n_colors=3)

fig, ax = plt.subplots(1,1, figsize=(5,4))
sns.scatterplot(data=data,
                x='t_inc',
                y='% mobile_particles',
                hue='flask',
                # markersize=50,
                palette=pal)
legend = plt.legend(loc='upper right', fontsize=15)
# legend.set_title('Rep. #', prop={'size': 14})
plt.ylabel('% Mobile basal bodies', fontsize=18)
plt.xlabel('Time under induction (min)', fontsize=18)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15);

plt.xlim(left=0, right=25)
plt.ylim(bottom=0, top=40)

plt.savefig(join(fig_folder, fig_savename),
           dpi=450,
           transparent=True,
           bbox_inches='tight');
```
